### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

```python
# Passo a passo
# Passo 0 - entender a empresa e o desafio dela
# Passo 1 - importar base de dados
# Passo 2 - preparar a base de dados para a Int Artifical
    # 1o a IA precisa aprender, que é o Machine Learning. 
    # Precisa pegar uma base grande, mandar pra IA e assim ela entende e dá uma análise, com base no histórico
    # precisa que esteja num formato específico
# Passo 3 - treinar a IA 0> criar nosso modelo de análise: nota de crédito boa, ok, ruim
# Passo 4 - escolher qual o melhor modelo gerado no passo 3
# Passo 5 - usar o melhor modelo para fazer previsão de novos clientes

#----> pra isso tudo, usa pandas e scikit-learn

# Score de crédito:
    # Good = boa
    # Standard = ok
    # Poor = ruim
```

```python
# Passo 1 - importar base de dados
import pandas as pd

tabela = pd.read_csv('clientes.csv')
display(tabela)

```

```python
# Passo 2 - preparar a base de dados para a Int Artifical
    # 1o a IA precisa aprender, que é o Machine Learning. 
    # Precisa pegar uma base grande, mandar pra IA e assim ela entende e dá uma análise, com base no histórico
    # precisa que esteja num formato específico

display(tabela.info())

# aqui ela tá limpa e tratada, tá tudo completo.
# se precisasse de algo, seria igual a aula 2. drop, dropna, etc

# -- Dtype:
# int - número inteiro
# float - número com casa decimal
# object - texto

# ----IA NÃO CONSEGUE APRENDER / TREINAR COM COLUNAS DE TEXTO. PRECISA CONVERTER. Score tá ok, pq ele gerará essa inf
    # usar << LabelEncoder >> nas colunas como object
        # profissao
            # cientista, bombeiro, engenheiro, dentista, etc
            # pra cada valor desse, o LabelEncoder atribuirá um valor
             # cientista = 1, bombeiro = 2, engenheiro = 3, 4, 5, 6....
        # mix_credito
        # comportamento_pagamento

# LabelEncoder profissões
from sklearn.preprocessing import LabelEncoder  
#--> é a scikit learn, mas o manual pra importar ela é sklearn. Orientam puxar só o que precisa dela
# oreoricessing, criacao de int art, selecao de modelo, etc. Então por isso recomanda puxar só o que precisa

codificador_profissao = LabelEncoder()
tabela['profissao'] = codificador_profissao.fit_transform(tabela['profissao'])
# o novo valor da coluna prof é igual ao antigo valor aplicando a codif profissao 

codificador_credito = LabelEncoder()
tabela['mix_credito'] = codificador_credito.fit_transform(tabela['mix_credito'])


codificador_pagamento = LabelEncoder()
tabela['comportamento_pagamento'] = codificador_pagamento.fit_transform(tabela['comportamento_pagamento'])

display(tabela.info())
# mostra de novo a tabela com as alterações

```

```python
# PASSO 2.5 - DIVISÃO DE DADOS TREINO & DADOS TESTE

# 2 divisões:
    # precisa identificar os DADOS DE TREINO e os que serão DADOS DE TESTE
        # não é interessante passar a base inteira pra ela aprender.
        # !!!!!  DEIXA SEPARADO UMA PARTE DA BASE PRA TESTAR SE ESTÁ FUNCIONANDO
    # quem serãos os clientes para previsão
        # A COLUNA QUE QUER PREVER, CHAMA DE Y
        # AS QUE SERÃO CRUZADAS PARA ANALISAR, SÃO CHAMADAS DE X

# y --> é a coluna da base da dados que quero prever
# previsão geralmente é única, não é bom criar uma pra analisar mais que uma inf
y = tabela['score_credito']

# x --> são as colunas da base de dados que serão usadas para fazer a previsão
    # pode listar as que vc quer OU colocar todas, usando drop columns
# --- x = tabela.drop(columns='score_credito')
    # assim é só pra tirar uma
    # pode excluir outras, pra reduzir a qtd de inf desnecessárias e gerar análise inúteis. add lista em []
x = tabela.drop(columns=['score_credito', 'id_cliente'])


# depois de definido, separa base treino e base teste
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y)
    # por padrão é 75% treino e 25% teste. Normalmente de 20 a 40 teste, 80 a 60 treino.
    # se quiser, muda train_test_split pra (x,y,test_size=0.3)
    # SEMPRE MAIS DADOS PRA TREINO, E MENOS INF PRA TESTE
    
```

```python
# Passo 3 - treinar a IA 0> 
# criar nosso modelo de análise: nota de crédito boa, ok, ruim
# criar 2, treina 2, testa 2, compara elas e escolhe qual a melhor
# em algumas situações uma é melhor que a outra, por isso precisa treinar as 2
    # modelo 1: Árvor de Decisão: RandomForest
    # modelo 2: Vizinhos Próximos: Neirest Neighbors - KNN

# são 3 etapas para criar uma Int Art
    # 1° = importar IA
    # 2° = criar a IA
    # 3° = treinar a IA

# importa IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# criar IA 
modelo_arvorededecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# treinar IA
modelo_arvorededecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)

```

```python
# Passo 4 - escolher qual o melhor modelo gerado no passo 3

# cria predição dos modelos
previsao_arvoredecisao = modelo_arvorededecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

# depois disso, calcula a precisão dos 2 testes
from sklearn.metrics import accuracy_score
display(accuracy_score(y_teste, previsao_arvoredecisao))
display(accuracy_score(y_teste, previsao_knn))


```

```python
# Passo 5 - usar o melhor modelo para fazer previsão de novos clientes
# melhor modelo é o Árvore de Decisão (modelo_arvorededecisao), com 83,08% de precisão
# com isso, vc cria uma análise para futuros clientes

# --- IMPORTAR NOVOS CLIENTES PRA PREVER
tabela_novosclientes = pd.read_csv('novos_clientes.csv')
display(tabela_novosclientes)

# --- TRATA A TABELA DOS NOVOS CLIENTES
# na nova, a profissão e outros pode estar como texto. Precisa passar pra IA que foi feito isso.
# usar os mesmos codificadores na nova base, pra ajustar as colunas
# COPIA O CÓDIGO DE ANTES QUE FUNCIONA. Mas não precisa mais do codificador, pq ele já foi usado antes E TIRA O FIT, só usa transform
# ou seja, remove o codificador, altera o nome das variáveis e tira o fit, deixando só transform

# codificador_profissao = LabelEncoder()
    # remove codificador
# tabela['profissao'] = codificador_profissao.fit_transform(tabela['profissao'])
    # corrige variável começo e fim
    # altera de fit.transform só pra transform

tabela_novosclientes['profissao'] = codificador_profissao.transform(tabela_novosclientes['profissao'])

tabela_novosclientes['mix_credito'] = codificador_credito.transform(tabela_novosclientes['mix_credito'])

tabela_novosclientes['comportamento_pagamento'] = codificador_pagamento.transform(tabela_novosclientes['comportamento_pagamento'])


display(tabela_novosclientes)

# --- FAZ NOVA PREVISÃO COM MODELO MELHOR
nova_previsao = modelo_arvorededecisao.predict(tabela_novosclientes)
display(nova_previsao)
```
